```python
artists = pd.read_csv('/home/mdbruchard/code/dmassonfr/Spotify_hit_predictor/raw_data/artists.csv')
tracks = pd.read_csv('/home/mdbruchard/code/dmassonfr/Spotify_hit_predictor/raw_data/tracks.csv')
kaggle = pd.read_csv('/home/mdbruchard/code/dmassonfr/Spotify_hit_predictor/raw_data/kaggle_df.csv')
```

```python
first_artist = kaggle[['first_artist']]
```

```python
tracks.tail()
```

```python
artists.head()
```

```python
tracks['id_artists']
```

```python
artists.columns

artists.rename(columns={'popularity': 'artist_popularity',
                        'id':'id_artists',
                       }, inplace=True)
```

```python
tracks = pd.concat([tracks, first_artist], axis=1)
```

```python
tracks.drop(columns='id_artists', inplace=True)
```

```python
tracks.rename(columns={'first_artist': 'id_artists'}, inplace=True)
```

```python
tracks.head()
```

```python
artists.drop(columns='name', inplace=True)
```

```python
tracks = tracks.merge(artists, on='id_artists')
```

```python
tracks.head()
```

```python
for col in new_songs.columns:
    if col not in tracks.columns:
        print(col)
```

```python
new_songs.rename(columns={'genre': 'genres'}, inplace=True)
```

```python
tracks.drop(columns='time_signature', inplace=True)
```

```python
data = tracks.merge(new_songs, on=[col for col in tracks.columns], how='outer')
```

```python
data.drop(data[(data['speechiness'] > 0.9) & (data['energy'] < 0.3)].index, inplace=True)
```

```python
data['artists'] = data['artists'].apply(lambda x: x.strip("['']"))
```

```python
data['genres'] = data['genres'].dropna().apply(lambda x: x.strip("['']"))
```
